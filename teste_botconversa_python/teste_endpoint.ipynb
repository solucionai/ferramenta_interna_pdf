```python
import requests

def fetch_data(numero_wpp, base_url='http://localhost:8000'):
    """
    Fetch data from the FastAPI endpoint using the given numero_wpp.
    
    Args:
        numero_wpp (str): The 'numero_wpp' to search for in the database.
        base_url (str): The base URL of the FastAPI server. Default is 'http://localhost:8000'.
    
    Returns:
        dict: The response data from the server.
    """
    url = f"{base_url}/retrieve/{numero_wpp}"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError if the HTTP request returned an unsuccessful status code
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
data = fetch_data('+5567909636005')
display(data)
```

```python
import requests

def fetch_all_data(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            return data
        else:
            print(f"Error: {response.status_code}, Detail: {response.json()['detail']}")
    except requests.RequestException as e:
        print(f"Failed to make the request: {e}")

# URL do endpoint
url = "https://web-production-c353.up.railway.app/retrieve_all"

# Chamando a função
all_data = fetch_all_data(url)
print(all_data)
```

### limpar coleção

```python
import pymongo

# Função para limpar a coleção
def limpar_colecao(db_name, collection_name, mongo_uri='mongodb://localhost:27017/'):
    # Conectar ao MongoDB
    client = pymongo.MongoClient(mongo_uri)
    db = client[db_name]
    collection = db[collection_name]
    
    # Remover todos os documentos da coleção
    result = collection.delete_many({})
    
    # Informar quantos documentos foram removidos
    print(f"{result.deleted_count} documentos removidos da coleção '{collection_name}'.")

# Nome do banco de dados e da coleção
db_name = 'seu_banco_de_dados'
collection_name = 'sua_colecao'

# Executar a função para limpar a coleção
limpar_colecao(db_name, collection_name)
```
